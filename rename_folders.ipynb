```python
import os
translate = {"cane": "dog", "cavallo": "horse", "elefante": "elephant", "farfalla": "butterfly", "gallina": "chicken", "gatto": "cat", "mucca": "cow", "pecora": "sheep", "scoiattolo": "squirrel", "dog": "cane", "cavallo": "horse", "elephant" : "elefante", "butterfly": "farfalla", "chicken": "gallina", "cat": "gatto", "cow": "mucca", "spider": "ragno", "squirrel": "scoiattolo"}
```

```python
img_folder = r"C:\Users\Şebnem\Desktop\animals\raw-img"
```

```python
for dirname in os.listdir(img_folder):
    old_path = os.path.join(img_folder, dirname)

    if os.path.isdir(old_path):
        if dirname in translate:
            new_name = translate[dirname]
            new_path = os.path.join(img_folder, new_name)

            if not os.path.exists(new_path):
                os.rename(old_path, new_path)
                print(f"✅ {dirname} → {new_name}")
            else:
                print(f"⚠️ {new_name} zaten var, {dirname} atlandı.")
```
